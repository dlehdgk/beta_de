```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def eos(z, w0, wa, beta):
    if abs(beta) < 0.001:
        w = w0 + wa * np.log(1 + z)
    else:
        w = w0 + wa * ((1 + z) ** (-beta) - 1) / beta
    return w
```

```python
z = np.linspace(0, 1000, 200)
w0 = -0.847
wa = -0.96
beta = [0, 1, 2, 10]

for b in beta:
    plt.plot(z, eos(z, w0, wa, b), label=f"$\\beta=${b}")
plt.plot(z, eos(z, -1, 0, 0), label="$\\Lambda$")
plt.xlabel("Redshift")
plt.ylabel("w")
plt.xscale("log")
plt.legend()
plt.grid()
plt.show()
```
